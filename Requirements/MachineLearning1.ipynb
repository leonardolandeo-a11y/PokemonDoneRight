```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.cluster import KMeans
```

```python
Data = pd.read_csv("smogon.csv")
DataMoves = Data["moves"].copy()
```

```python
TextToNumbers = TfidfVectorizer(ngram_range=(1,1))
MatrixTextoToNumbers = TextToNumbers.fit_transform(DataMoves)
```

```python
MatrixTextoToNumbers.shape[1]
```

```python
Names = TextToNumbers.get_feature_names_out()
indice = 0
for i in TextToNumbers.vocabulary_:
    indice +=1
indice
```

```python
MatrixTextoNumbersDataFrame = pd.DataFrame(MatrixTextoToNumbers.toarray(),columns=Names)
Cluster = KMeans(n_clusters=18,random_state=42)
ClustersPokemon = Cluster.fit(MatrixTextoNumbersDataFrame)
MatrixTextoNumbersDataFrame["Clusters"] = ClustersPokemon.labels_
MatrixTextoNumbersDataFrame.insert(0,"NombrePokemon",Data["Pokemon"] )
MatrixTextoNumbersDataFrame.to_csv("PokemonDataSet.csv",index = False )
```
